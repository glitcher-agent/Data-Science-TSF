## **To Explore Unsupervised Machine Learning**

Importing datasets

```
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

Loading Data

```
ir=datasets.load_iris()
iris_df = pd.DataFrame(ir.data, columns = ir.feature_names)
iris_df.head()

```

Target and Predictors

```
X = ir.data[:, :2]
y = ir.target
print(X)
print(y)
```

```
plt.scatter(X[:,0], X[:,1], c=y, cmap='Dark2')
plt.xlabel('Spea1 Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

# **Instantiate and fit our K means cluster model**

```
km = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
km.fit(X)
```

```
centers = km.cluster_centers_
print(centers)
```

# **Plot the identified clusters and compare with the answers**

```
new = km.labels_
fig, axes = plt.subplots(1, 2, figsize=(20,5))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='Dark2')
axes[1].scatter(X[:, 0], X[:, 1], c=new, cmap='gist_rainbow')
axes[0].set_xlabel('Sepal length', fontsize=20)
axes[0].set_ylabel('Sepal width', fontsize=20)
axes[1].set_xlabel('Sepal length', fontsize=20)
axes[1].set_ylabel('Sepal width', fontsize=20)
axes[0].set_title('Actual', fontsize=20)
axes[1].set_title('Predicted', fontsize=20)
```
